```python
import schemdraw
import schemdraw.elements as elm
```

```python
def create_C0_L0(d, name):
    d += elm.Dot().label('', 'top', fontsize=20, font='serif')
    d += elm.Capacitor().label('$C_c$', fontsize=20, font='serif').down()
    d += elm.Line().down().length(0.5)
    d += elm.Line().left().length(1.2)
    d += elm.Capacitor().label('$C$', fontsize=20, font='serif').down()
    d += elm.Line().right(1.2)
    d += elm.Ground()
    d += elm.Line().right(1.2)
    d += elm.Inductor2().label('$L$', fontsize=20, font='serif').up()
    d += elm.Line().left(1.2)
    d.push()
    d.pop()
```

```python
with schemdraw.Drawing() as d:
    d += elm.Dot().label('1', 'left', fontsize=20, font='serif')
    coax = elm.cables.Coax()
    # d += elm.cables.Coax().label('$V_{in}$', fontsize=20, font='serif')
    create_C0_L0(d, 'C0_L0')
    d += elm.cables.Coax().at(coax.end).right()
    d += elm.Dot().label('2', 'right', fontsize=20, font='serif')
    d.pop()
    d.save(r'C:\Users\LocalAdmin\OneDrive - Delft University of Technology\gonogo-pics\hanged-resonator.png')

```

# Cheby

```python
def create_ground_C(d, name):
    d += elm.Dot()
    # d.push()
    # d += elm.Line().down().length(0.5)
    d.push()
    d += elm.Capacitor().label(name, fontsize=20, font='serif').down()
    d += elm.Dot()
    d.pop()
```

```python
def create_ground_C_not_push(d, name):
    d += elm.Dot()
    # d.push()
    # d += elm.Line().down().length(0.5)

    d += elm.Capacitor().label(name, fontsize=20, font='serif').down()
    d += elm.Dot()
    d.pop()
```

```python
with schemdraw.Drawing() as d:
    d+=elm.Dot(open=True)
    d += elm.Line().right().length(1.5)
    create_ground_C(d, '$C_1$')
    d += elm.Inductor().label('$L_2$', fontsize=20, font='serif').length(4)
    create_ground_C(d, '$C_3$')
    d += elm.Inductor().label('$L_4$', fontsize=20, font='serif').length(4)
    create_ground_C_not_push(d, '$C_5$')
    d += elm.Line().left().length(9.5)
    d+=elm.Dot(open=True)
    d.save('cheby_circuit.png')

    
```
