```python

import numpy as np
import matplotlib.pyplot as plt
from Schuster_Resonators.approx_Schuster import *
from CPW.cpw import *
import scipy.io as sc
import scipy.signal as sig
```

```python
path = r'C:\Users\localadmin\Documents\GitHub\resonators\mat_files\20241001-GeRes-Pt-13pHsq'
file_list = os.listdir(path)
print(file_list)
d = {}

# Load the data
for file in file_list:
    name = file.split('.')[0]
    file_path = os.path.join(path, file)
    mat_contents = sc.loadmat(file_path)
    freq = mat_contents['X\x00\x00\x00\x00\x00\x00\x00'][0]
    amp = mat_contents['Y\x00\x00\x00\x00\x00\x00\x00'][0]
    d[name] = {'freq': freq, 'amp': amp}
```

```python
freqs = []
for name in d.keys():
    plt.plot(d[name]['freq'], d[name]['amp'])
    peak, _ = sig.find_peaks(-d[name]['amp'], height=0.4)
    plt.plot(d[name]['freq'][peak], d[name]['amp'][peak], 'rx')
    for p in peak:
        freqs.append(1e-9*d[name]['freq'][p])
plt.xlabel('Frequency (GHz)')
plt.ylabel('S21 (dB)')
plt.title('GeRes Pt 13pHsq')
```
