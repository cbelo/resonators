```python
import numpy as np
import scipy.constants as const
import matplotlib.pyplot as plt
```

$L_{k,sq} = \frac{R_s h}{2π^2 Δ}\cdot \frac{1}{\mathrm{tanh}⁡(Δ/2kT)} \approx \frac{R_s h}{2π^2\cdot 1.76kT_c} = a \cdot\frac{R_s}{T_c}$

```python
a = const.h /(2*np.pi**2*1.76*const.k)
a
```

```python
Rs_20nmTUD = np.mean(np.array([68.64259435, 79.22565528, 77.54868203, 77.48825056, 77.05012241, 77.15587748, 77.42781909, 76.95947521, 77.66954496]))
Rs_SC = 149.9
Rs_chipsCPH = 1e3*np.mean(np.array([1.350643322, 1.349283614, 1.337046242, 1.341578602]))
Rs_HarryGe = np.mean(np.array([329.3666115, 336.7543585, 332.2219984, 331.3155264]))

Tc = np.linspace(7,11,100)
Lk_20nmTUD = 1e12*a*Rs_20nmTUD/Tc
Lk_SC = 1e12*a*Rs_SC/Tc
Lk_chipsCPH = 1e12*a*Rs_chipsCPH/Tc
Lk_HarryGe = 1e12*a*Rs_HarryGe/Tc
plt.plot(Tc, Lk_20nmTUD, label = "SiTUD20")
plt.plot(Tc, Lk_SC, label = "SiSC10")
plt.plot(Tc, Lk_chipsCPH, label = "SiChip5")
plt.plot(Tc, Lk_HarryGe, label = "GeHarry8")
plt.xlabel("$T_c$ (K)")
plt.ylabel("$L_k$ (pH/sq)")
plt.plot(9, 1e12*a*Rs_20nmTUD/9, 'bo')
plt.plot(8.5, 1e12*a*Rs_SC/8.5, 'o', color = 'orange')
plt.plot(8, 1e12*a*Rs_chipsCPH/8, 'go')
plt.plot(7.5, 1e12*a*Rs_HarryGe/7.5, 'ro')
plt.legend()
print('SiTUD20',1e12*a*Rs_20nmTUD/9)
print('SiSC10',1e12*a*Rs_SC/8.5)
print('SiChip5',1e12*a*Rs_chipsCPH/8)
print('GeHarry8', 1e12*a*Rs_HarryGe/7.5)
```
